```python
import pathlib
import tifffile as tiff


def generate_positions(length, tile_size, overlap):
    """ 
    Gets tile starting position while including overlap.
    
    ___PARAMETERS___
    length     = total length of image axis (int)
    tile_size  = desired tile size (int)
    overlap    = size of overlap (int)
    """
    #Genererates tile coordinates
    stride = tile_size - overlap
    positions = list(range(0, max(length - tile_size + 1, 1), stride))

    #adding final tile coordinate (if needed)
    last_pos = length - tile_size
    if last_pos < 0:
        last_pos = 0

    if not positions or positions[-1] != last_pos:
        positions.append(last_pos)

    return positions


def tile_tiff(input_path, output_directory, tile_height, tile_width, overlap):
    """
    Processes image into tiles.
    
    ___PARAMETERS___
    input_path       = name of image file for tiling (str)
    output_directory = name of output directory (str)
    tile_height      = desired tile height (int)
    tile_width       = desired tile width (int)
    overlap          = desired overlap (int)
    """  
    #Preprocessing 
    image = tiff.imread(input_path)
    height, width = image.shape[:2]
    y_positions = generate_positions(height, tile_height, overlap)
    x_positions = generate_positions(width, tile_width, overlap)

    #Output map generation
    output_directory = pathlib.Path(output_directory)
    output_directory.mkdir(parents=True, exist_ok=True)

    #Tile processing
    tile_count = 0
    for y in y_positions:
        for x in x_positions:
            #Generates image
            tile = image[y:y + tile_height, x:x + tile_width]
            
            #Stores image
            out_name = f"tile_y{y}_x{x}.tif"
            out_path = output_directory / out_name
            tiff.imwrite(out_path, tile)
            
            tile_count += 1
    print(f"Number of tiles in {output_directory}: {tile_count} tiles")


#Usage
tile_tiff(input_path = "25FU1231.tif", 
          output_directory = "Example_folder", 
          tile_height = 512, 
          tile_width = 512, 
          overlap=64)
```

```python
import pathlib
import tifffile as tiff
from concurrent.futures import ThreadPoolExecutor


def generate_positions(length, tile_size, overlap):
    """
    Gets tile starting position while including overlap.

    ___PARAMETERS___
    length     = total length of image axis (int)
    tile_size  = desired tile size (int)
    overlap    = size of overlap (int)
    """
    # Generates tile coordinates
    stride = tile_size - overlap
    positions = list(range(0, max(length - tile_size + 1, 1), stride))

    # Adding final tile coordinate (if needed)
    last_pos = length - tile_size
    if last_pos < 0:
        last_pos = 0

    if not positions or positions[-1] != last_pos:
        positions.append(last_pos)

    return positions


def save_tile(image, y, x, tile_height, tile_width, output_directory):
    """
    Saves a single tile.
    """
    # Generate image tile
    tile = image[y:y + tile_height, x:x + tile_width]

    # Store image
    out_name = f"tile_y{y}_x{x}.tif"
    out_path = output_directory / out_name
    tiff.imwrite(out_path, tile)

    return out_name


def tile_tiff(input_path, output_directory, tile_height, tile_width, overlap):
    """
    Processes image into tiles using ThreadPoolExecutor.

    ___PARAMETERS___
    input_path       = name of image file for tiling (str)
    output_directory = name of output directory (str)
    tile_height      = desired tile height (int)
    tile_width       = desired tile width (int)
    overlap          = desired overlap (int)
    """

    # Preprocessing
    image = tiff.imread(input_path)
    height, width = image.shape[:2]

    y_positions = generate_positions(height, tile_height, overlap)
    x_positions = generate_positions(width, tile_width, overlap)

    # Output folder generation
    output_directory = pathlib.Path(output_directory)
    output_directory.mkdir(parents=True, exist_ok=True)

    # Create all tile jobs
    jobs = []
    for y in y_positions:
        for x in x_positions:
            jobs.append((y, x))

    # Run with 8 threads
    with ThreadPoolExecutor(max_workers=8) as executor:
        futures = [
            executor.submit(
                save_tile,
                image,
                y,
                x,
                tile_height,
                tile_width,
                output_directory
            )
            for y, x in jobs
        ]

        # Wait for completion
        for future in futures:
            future.result()

    print(f"Number of tiles in {output_directory}: {len(jobs)} tiles")


# Usage
tile_tiff(
    input_path="25FU1231.tif",
    output_directory="Example_folder",
    tile_height=512,
    tile_width=512,
    overlap=64
)
```
